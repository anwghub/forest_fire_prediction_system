```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
df = pd.read_csv("forest_fire_dataset.csv")
```

```python
df.head()
```

```python
le_month = LabelEncoder()
le_day = LabelEncoder()
```

```python
df['month'] = le_month.fit_transform(df['month'])
df['day'] = le_day.fit_transform(df['day'])
```

```python
df['fire_occurred'] = df['area'].apply(lambda x: 1 if x > 0 else 0)
```

```python
X_class = df.drop(['area', 'fire_occurred'], axis=1)
y_class = df['fire_occurred']
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(Xc_train, yc_train)
```

```python
df_fire_only = df[df['fire_occurred'] == 1]

X_reg = df_fire_only.drop(['area', 'fire_occurred'], axis=1)
y_reg = df_fire_only['area']
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(Xr_train, yr_train)
```

```python
joblib.dump(regressor, 'regressor.pkl')
joblib.dump(classifier, 'classifier.pkl')
joblib.dump(le_month, 'le_month.pkl')
joblib.dump(le_day, 'le_day.pkl')
```
